### Diezmado e interpolación
Solamente vamos a mostrar cómo realizar estos procesos y qué efecto tienen

```python
import scipy.signal as sig
import numpy as np
import matplotlib.pyplot as plt
from scipy.fftpack import fft, ifft, fftshift, fftfreq
from __future__ import division

########### Parámetros Iniciales ###########
Fs = 10000  # Sampling Freq 10KHz
N = 100      # Samples number
t = np.linspace(0,N/Fs,num=N) #Crea una variable de tiempo desde 0, con intervalos de 1/Fs

f1 = 100   # 100Hz
A1 = 5
fx= fftshift(fftfreq(N,1/Fs))

FACTOR_DIEZMADO = 4

signal_1 = A1 * np.sin(2*np.pi*f1*t)

plt.title("Original")
plt.stem(t,signal_1)
plt.show()
#Obtener y graficar el espectro

spec_1 = abs(fftshift(fft(signal_1)))
plt.title("Espectro Original")
plt.stem(fx,spec_1)
```

```python
signal_2 = sig.decimate(signal_1,FACTOR_DIEZMADO) #Esta función aplica un filtro pasa bajos (chebyshev 8.bit) y luego diezma 
t_2 = sig.decimate(t,FACTOR_DIEZMADO) #Si uso un array como base de tiempo también tiene que diezmarse (obviamente)
plt.title("Senal diezmada") #no admite la ENIE >:c
plt.stem(t_2,signal_2)
plt.show()

spec_2 = abs(fftshift(fft(signal_2)))
plt.title("Espectro tras diezmado")
fx_2 = sig.decimate(fx,FACTOR_DIEZMADO)
plt.stem(fx_2, spec_2)
```
